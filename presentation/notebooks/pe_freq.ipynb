```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
# for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
# t = [0, 0.3, 0.5, 0.7]
t = 0.5

def f(x, freq=0.1):
    return np.sin(x * freq)
```

```python
xrange = np.linspace(4850, 4950, 100)
v=4920
v2=4910
f1=5
f2=15
plt.figure(figsize=(4,2), dpi=300)
plt.axvline(v, linestyle='--', linewidth=1, color='k')
plt.plot(xrange, f(xrange, freq=1/f1), label='freq\\_1', linestyle='-', color='k')
plt.plot(xrange, f(xrange, freq=1/f2), label='freq\\_2', linestyle=':', color='k')
plt.scatter(v, f(v, freq=1/f1), marker='s', color='k')
plt.scatter(v, f(v, freq=1/f2), marker='s', color='k')
# plt.scatter(v2, f(v2, freq=1/f1), marker='s', color='k')
# plt.scatter(v2, f(v2, freq=1/f2), marker='s', color='k')
plt.ylabel('PE encoded value')
plt.xlabel('Observation time [MJD]')
leg = plt.legend(bbox_to_anchor=(0.85, 1.23), ncol=2)
leg.get_frame().set_edgecolor('k')
plt.savefig('../../presentation/figures/pe_freq.pdf', format='pdf', bbox_inches='tight')
```
